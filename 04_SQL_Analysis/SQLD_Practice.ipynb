```python
import pandas as pd
import sqlite3

conn = sqlite3.connect(':memory:')

# 1. TAB1 (ì™¼ìª½ í…Œì´ë¸”): ê¸°ì¤€ì´ ë˜ëŠ” í…Œì´ë¸”
# - íŠ¹ì§•: (1, A)ëŠ” TAB2ì— 2ê°œ ìˆìŒ (1:N)
# - íŠ¹ì§•: (2, B)ëŠ” TAB2ì— ì—†ìŒ (ë§¤ì¹­ ì‹¤íŒ¨)
# - íŠ¹ì§•: (NULL, C)ëŠ” COL1ì´ NULLì„
data1 = {
    'COL1': [1, 2, None], 
    'COL2': ['A', 'B', 'C']
}

# 2. TAB2 (ì˜¤ë¥¸ìª½ í…Œì´ë¸”)
# - íŠ¹ì§•: (1, A)ê°€ 2ê°œ ìˆìŒ (ì¤‘ë³µ ë°ì´í„°)
data2 = {
    'COL1': [1, 1], 
    'COL2': ['A', 'A']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df1.to_sql('TAB1', conn, index=False)
df2.to_sql('TAB2', conn, index=False)

# -------------------------------------------------------
# 3. ì¤‘ê°„ ê³¼ì • í™•ì¸ (JOIN ê²°ê³¼ê°€ ì–´ë–»ê²Œ ìƒê²¼ë‚˜?)
# -------------------------------------------------------
print("=== 1ë‹¨ê³„: LEFT JOIN ê²°ê³¼ (ëˆˆìœ¼ë¡œ í™•ì¸) ===")
print(pd.read_sql("""
    SELECT 
        TAB1.COL1 AS T1_COL1, 
        TAB1.COL2 AS T1_COL2,
        TAB2.COL1 AS T2_COL1, 
        TAB2.COL2 AS T2_COL2
    FROM TAB1 
    LEFT OUTER JOIN TAB2 
    ON TAB1.COL2 = TAB2.COL2 
    AND TAB1.COL1 = TAB2.COL1
""", conn))

# -------------------------------------------------------
# 4. ìµœì¢… ë¬¸ì œ í’€ì´ (COUNT)
# -------------------------------------------------------
print("\n=== 2ë‹¨ê³„: ìµœì¢… COUNT ê²°ê³¼ ===")
print(pd.read_sql("""
    SELECT COUNT(TAB1.COL1) AS CNT
    FROM TAB1 
    LEFT OUTER JOIN TAB2 
    ON TAB1.COL2 = TAB2.COL2 
    AND TAB1.COL1 = TAB2.COL1
""", conn))
```
