#### 2. JOINì˜ ì¢…ë¥˜ (ë²¤ ë‹¤ì´ì–´ê·¸ë¨ìœ¼ë¡œ ì´í•´í•˜ê¸°)

```python
import pandas as pd
import sqlite3

# 1. ê°€ìƒì˜ DB ë§Œë“¤ê¸°
conn = sqlite3.connect(':memory:')

# 2. í…Œì´ë¸” ë°ì´í„° ìƒì„± (SQLD ê¸°ì¶œ ìœ í˜•)
# EMP(ì§ì›) í…Œì´ë¸”
emp_data = {
    'EMP_ID': [100, 101, 102, 103, 104],
    'NAME': ['ì² ìˆ˜', 'ì˜í¬', 'ë¯¼ìˆ˜', 'ì§€ìˆ˜', 'ë™í˜„'],
    'DEPT_ID': [10, 10, 20, None, 30], # ì§€ìˆ˜ëŠ” ë¶€ì„œê°€ ì—†ìŒ(NULL)
    'SALARY': [3000, 3500, 4000, 2500, 4500]
}
# DEPT(ë¶€ì„œ) í…Œì´ë¸”
dept_data = {
    'DEPT_ID': [10, 20, 40], # 30ë²ˆ ë¶€ì„œëŠ” ì—¬ê¸° ì—†ìŒ
    'DEPT_NAME': ['ì¸ì‚¬íŒ€', 'ê°œë°œíŒ€', 'ê¸°íšíŒ€']
}

df_emp = pd.DataFrame(emp_data)
df_dept = pd.DataFrame(dept_data)

# DBì— ì €ì¥
df_emp.to_sql('EMP', conn, index=False)
df_dept.to_sql('DEPT', conn, index=False)

def run_query(q):
    return pd.read_sql(q, conn)

# -------------------------------------------------------
# Q1. INNER JOIN vs LEFT JOIN ì°¨ì´ í™•ì¸í•˜ê¸°
# -------------------------------------------------------
print("1. INNER JOIN ê²°ê³¼ (êµì§‘í•©):")
# ë¬¸ì œ: ë¶€ì„œê°€ ë°°ì •ëœ ì§ì›ë§Œ ë‚˜ì˜µë‹ˆë‹¤. (ì§€ìˆ˜ ì œì™¸, 30ë²ˆ ë¶€ì„œ ë™í˜„ ì œì™¸)
print(run_query("""
    SELECT A.NAME, B.DEPT_NAME
    FROM EMP A
    INNER JOIN DEPT B ON A.DEPT_ID = B.DEPT_ID
"""))

print("\n2. LEFT JOIN ê²°ê³¼ (ì „ì²´ ì§ì›):")
# ë¬¸ì œ: ë¶€ì„œê°€ ì—†ëŠ” 'ì§€ìˆ˜'ì™€ ë¶€ì„œì •ë³´ì— ì—†ëŠ” ë¶€ì„œì¸ 'ë™í˜„'ì€ ì–´ë–»ê²Œ ë‚˜ì˜¬ê¹Œìš”?
print(run_query("""
    SELECT A.NAME, A.DEPT_ID, B.DEPT_NAME
    FROM EMP A
    LEFT JOIN DEPT B ON A.DEPT_ID = B.DEPT_ID
"""))

# -------------------------------------------------------
# Q2. NULL ì²˜ë¦¬ê°€ í¬í•¨ëœ ì§‘ê³„ (í•¨ì • ë¬¸ì œ)
# -------------------------------------------------------
print("\n3. COUNT í•¨ìˆ˜ì˜ í•¨ì •:")
# ë¬¸ì œ: ì „ì²´ í–‰ ê°œìˆ˜ vs ë¶€ì„œIDê°€ ìˆëŠ” í–‰ ê°œìˆ˜ ì°¨ì´
print(run_query("""
    SELECT 
        COUNT(*) as ì „ì²´í–‰ìˆ˜,
        COUNT(DEPT_ID) as ë¶€ì„œìˆëŠ”ì§ì›ìˆ˜,
        SUM(SALARY) as ê¸‰ì—¬í•©ê³„
    FROM EMP
"""))
```
