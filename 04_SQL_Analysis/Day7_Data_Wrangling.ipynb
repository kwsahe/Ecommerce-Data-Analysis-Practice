```python
import pandas as pd
import numpy as np

# 1. ì§€ì˜¥ì—ì„œ ì˜¨ ë”ëŸ¬ìš´ ë°ì´í„° ìƒì„± (ì‹¤ìŠµìš©)
# ìƒí™©: ì—‘ì…€ì—ì„œ ëŒ€ì¶© ì…ë ¥í•´ì„œ í˜•ì‹ì´ ì—‰ë§ì¸ ë°ì´í„°ê°€ ë„˜ì–´ì™”ë‹¤.
raw_data = {
    'order_id': ['O001', 'O002', 'O003', 'O004', 'O005'],
    'user_id': ['User_A', 'User_B', 'User_A', 'User_C', 'User_B'],
    'amount': ['50,000ì›', '30,000', '15,000ì›', '-', '100,000'], # ì½¤ë§ˆ, ì›, í•˜ì´í”ˆ(-) í˜¼ì…
    'date': ['2024.01.01', '2024-01-03', '24/01/05', '2024.01.07', '2023-12-25'], # ë‚ ì§œ í˜•ì‹ ì œê°ê°
}

df_dirty = pd.DataFrame(raw_data)

print("--- ğŸ˜± ì •ë¦¬ ì•ˆ ëœ ë°ì´í„° (Raw Data) ---")
print(df_dirty)
print("\n--- ë°ì´í„° íƒ€ì… í™•ì¸ ---")
print(df_dirty.info())
```

```python
df_dirty['amount'] = df_dirty['amount'].str.replace('ì›', '')

df_dirty['amount'] = df_dirty['amount'].str.replace(',', '')

df_dirty['amount'] = pd.to_numeric(df_dirty['amount'], errors='coerce')

df_dirty['amount'] = df_dirty['amount'].fillna(0)

print("--- âœ¨ ê¹¨ë—í•´ì§„ ê¸ˆì•¡ ë°ì´í„° ---")
print(df_dirty)
print(df_dirty.info())
```

```python
df_dirty['date'] = pd.to_datetime(df_dirty['date'], format='mixed')

df_dirty['year'] = df_dirty['date'].dt.year         
df_dirty['month'] = df_dirty['date'].dt.month       
df_dirty['day_name'] = df_dirty['date'].dt.day_name() # ìš”ì¼ ì´ë¦„ (Monday...)

print("--- ğŸ“… ë‚ ì§œ ë³€í™˜ ë° ìš”ì¼ ì¶”ì¶œ ì™„ë£Œ ---")
print(df_dirty)
print("\n--- ë°ì´í„° íƒ€ì… í™•ì¸ (datetime64ê°€ ë‚˜ì™€ì•¼ ì„±ê³µ) ---")
print(df_dirty.info())
```
